# **Testando o Ollama Local**

### Modelo utilizado

Este notebook utiliza o modelo **`phi3:latest`**, executado localmente através do Ollama.

O Phi-3 é um modelo leve e eficiente, ideal para testes, estudos e aplicações locais sem necessidade de conexão com APIs externas.

### Como instalar o Ollama

Siga os passos abaixo para configurar o ambiente:

#### **1. Instalar o Ollama**

Acesse o site oficial e faça o download:

https://ollama.com

Ou via terminal (Linux/macOS):

```
curl -fsSL https://ollama.com/install.sh | sh
```

#### **2. Baixar o modelo Phi-3**

Após instalar o Ollama, execute:

```
ollama pull phi3
```

#### **3. Iniciar o servidor**

O Ollama normalmente inicia automaticamente, mas você pode garantir com:

```
ollama serve
```

O servidor ficará disponível em:

```
http://localhost:11434
```

````python
# Importando as bibliotecas necessárias
import os
import requests
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from openai import OpenAI
````

````python
# Inicializa cliente OpenAI apontando para servidor local (Ollama)
openai = OpenAI(
    base_url='http://localhost:11434/v1', 
    api_key='ollama'
)
````

### **Fazendo uma chamada simples para ver se está funcionando**

````python
# Para dar uma prévia -- fazer a chamada à API com essas mensagens é assim tão fácil. Se tiver problemas, vá para o notebook de Solução de Problemas.

message = "Hello, LLM! This is my first message to you!"
response = openai.chat.completions.create(model="phi3:latest", messages=[{"role":"user", "content":message}])
print(response.choices[0].message.content)
````
